```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins').dropna()
penguins
```

```python
#penguins ë°ì´í„°ì— linplotì„ ì¶œë ¥(ì‹œê°„ì— ë”°ë¥¸ ë³€í™”ëž‘):ì‹ ë¢°ë„ ì œê±° ci= None,hueëŠ” ì¹´í…Œê³ ë¦¬ë³„ë¡œ ë‹¤ë¥¸ìƒ‰ê¹”
sns.lineplot(data=penguins,x="body_mass_g", y="flipper_length_mm",ci=None, hue='species')
```

```python
#pointplot(ìœ ì˜ë¯¸í•œ ë³€í™”ê´€ì°°) ìƒ‰ê¹”ë°”ê¾¸ê¸° palettte
sns.pointplot(data=penguins, x="species",y="body_mass_g" ,hue="sex",palette="Set2")
```

```python
#í¬ê¸°í‘œí˜„í•˜ê¸°
#penguin ë°ì´í„°ì— histplot ì¶œë ¥(ë§‰ëŒ€ê¸°ì„¸ìš°ê¸°),yì¶•ê³ ì •(ì£¼ì˜)ì¡°ê¸ˆë”ìžì„¸ížˆ bins,ìœ„ë¡œìŒ“ê¸° multiple='stack'
sns.histplot(data=penguins, x="bill_depth_mm",bins=30, hue="species",multiple='stack')
```

```python
#barplot (ë§‰ëŒ€ê·¸ëž˜í”„)
sns.barplot(data=penguins,x="island",y="body_mass_g" ,hue="sex")
```

```python
# countplot(ë²”ì£¼í˜• ì†ì„±ì„ ê°€ì§€ëŠ” ë°ì´í„°ë“¤ì˜ histogramì„ ë³´ì—¬ì£¼ëŠ” plot, ì¢…ë¥˜ë³„ countë¥¼ ë³´ì—¬ì£¼ëŠ” ë°©ë²•)
```
