```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins').dropna()
penguins
```

```python
#penguins 데이터에 linplot을 출력(시간에 따른 변화랑):신뢰도 제거 ci= None,hue는 카테고리별로 다른색깔
sns.lineplot(data=penguins,x="body_mass_g", y="flipper_length_mm",ci=None, hue='species')
```

```python
#pointplot(유의미한 변화관찰) 색깔바꾸기 palettte
sns.pointplot(data=penguins, x="species",y="body_mass_g" ,hue="sex",palette="Set2")
```

```python
#크기표현하기
#penguin 데이터에 histplot 출력(막대기세우기),y축고정(주의)조금더자세히 bins,위로쌓기 multiple='stack'
sns.histplot(data=penguins, x="bill_depth_mm",bins=30, hue="species",multiple='stack')
```

```python
#barplot (막대그래프)
sns.barplot(data=penguins,x="island",y="body_mass_g" ,hue="sex")
```

```python
# countplot(범주형 속성을 가지는 데이터들의 histogram을 보여주는 plot, 종류별 count를 보여주는 방법)
```

```python
#penguin 데이터에 countplot을 출력합니다.
sns.countplot(data=penguins, x="species",hue='sex')
```

## 분포를 표현하는 그래프
- boxplot, swarmplot, scatterpolot

```python
sns.boxplot(data=penguins, x="species", y="body_mass_g",hue='sex')
```

```python
#swarmplot 데이터에 대한 분포 자체를 보여주는 plot
sns.swarmplot(data=penguins, x="sex",y="flipper_length_mm",hue='sex')
```

```python
#scatterplot linplot과 비슷하게 x,y에 대한 전체적인 분포를 확인하는 plot
#lineplot은 경향성에 초점을 둔다면, scatterplot은 데이터 그자체에 퍼져있는 모양에 중점을 둔다
#상관관계를 볼수있다(양의상관관계)

sns.scatterplot(data=penguins, x="body_mass_g", y="flipper_length_mm",hue='species')
```

## 연관성파악하기
- heatmap
- pairplot

```python
# 각 colum과 상관관계를 파악하기 위해 coreelation matrix를 만든다.
penguins.corr()
```

```python
#penguin 데이터에 heatmap을 출력한다.(수치값표현 annot, 색깔다루기CMAP)
sns.heatmap(data=penguins.corr(), annot=True,cmap="Blues")
```

```python
#pairplot
sns.pairplot(data=penguins,hue="species")
```
