```python
#태그에 있는 속성값자체가 필요한 경우 딕셔너리 인덱싱처럼 들고온다 
last_url = sp.find_all('td', class_="pgRR")[0].find_all("a")[0]["href"]
```

```python
#마지막 글자 떼어내기 위해 규칙을 생각해본다.
last_page=int(last_url.split('&page=')[-1])
```

```python
last_page
```

```python
date_list = []
price_list=[]





for pageno in range(1,last_page+1):
  page_url = f"https://finance.naver.com/sise/sise_index_day.naver?code=KOSPI&page={pageno}"
  source = requests.get(page_url)
  sp = BeautifulSoup(source.content,'html.parser')
  dates = sp.find_all('td', class_='date')
  
  for date in dates:
    date_list.append(date.text)
    
  prices = sp.find_all("td", class_="number_1")
    
  for price in prices[::4]:
    price_list.append(price.text)
```

```python
date_list
```

```python
len(date_list), len(price_list)
```

```python
#파일저장 코드
import pandas as pd
df = pd.DataFrame({"date":date_list, "price":price_list})

df.to_excel("g:/kpiindex.xlsx", index=False)
```
