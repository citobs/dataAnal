```python
import pandas as pd
from glob import glob
from tqdm.notebook import tqdm
import os

#특정경로에 있는 파일이름을 가져온다 glob
```

```python
temp = pd.read_excel('../crawldata/지역_위치별(주유소) (1).xls')
temp

#데이터 전처리
temp = pd.read_excel('../crawldata/지역_위치별(주유소) (1).xls', header=2)
temp
```

```python
#특정경로에 있는 파일이름으로 가져온다(규칙 확장자가 xls인것)
#file붙이기함수는 concat
station_files = glob('../crawldata/*.xls')
station_files

total = pd.DataFrame()
for file_name in station_files:
    temp = pd.read_excel(file_name, header=2)
    
    #세로로 이어붙이기[temp,temp,~~]
    total = pd.concat([temp,temp])

total = total.sort_values(by="지역")
#번호(인덱스)날리고 새로 부여
total = total.reset_index(drop=True)
total

```

```python
#저장코드
total.to_excel('g:/서울시주유소가격.xlsx')
```
